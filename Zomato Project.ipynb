```python
#convert the data type of column-rate
def handleRate(value):
    value=str(value).split('/')
    value= value[0];
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

```python
# Type of resturant
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("type of resturant")
```

```python
#conclusion --Majority of the resturant falls in dinning category
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data}).reset_index()

plt.plot(result['listed_in(type)'], result['votes'], c="red", marker="o")
plt.xlabel("Type of restaurant", color="red", size=20)
plt.ylabel("Votes", color="red", size=20)
plt.xticks(rotation=90)  # To ensure labels are readable if there are many categories
plt.show()
```

```python
#conclusion-- dinning resturants has recieved maximum votes
```

```python
plt.hist(dataframe['rate'],bins=5)
plt.title("ratings distribution")
plt.show()
```

```python
#conclusion -- the majority resturants received ratings form 3.5 to 4
```

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

```python
# the majority of couples preferr resturants with 300  rupes
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=dataframe)
```

```python
#conslusion-- ofline order received lower rating in comparison to online order
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns="online_order", fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')  # '.1f' for one decimal place
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

```python


# Assuming 'dataframe' is defined elsewhere
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns="online_order", aggfunc='size', fill_value=0)

# Create heatmap with similar style as the image
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')

# Add title and labels
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")

# Display the heatmap
plt.show()
```
